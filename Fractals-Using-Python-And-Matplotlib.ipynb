<font size="3">Fractals are complex and repeating geometrical patterns that are self-similar accross different scales and can be found in nature in the form of leaves or flowers while more complex fractals can be generated by solving certain mathematical equations. Most online tutorials on plotting fractals using python use the Turtle library. In this article, I will be using only matplotlib and python to draw the Koch Snowflake and Sierpinski triangle. But the code can be modified and can be used to plot other fractals as long as one knows what patterns to use. The code will have two parts. The first, drawFractal(), to translate the given pattern and store x & y values in lists and use those lists to draw the fractal. The second part will define the fractals themselves according to a given initial pattern and the rule for repetition of the pattern. </font?
    
<br><br>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<br><b> <font size="3"> Function to be called to draw the fractals </font> </b>
<br><br><font size="3">
Since fractals are repeating patterns, the building blocks of the pattern must be interpreted and registered so that the data can be used to plot the fractal. In this case, a forward movement and rotation angle are used to make up the overall pattern of the fractal. Forward movement is represented by F and angle of rotation is represented by - and + for clockwise and counterclockwise respectively. Sometimes to differentiate one forward movement (horizontal) from the forward movement after rotation (vertical), the letter G may be used to for vertical movement. For simplicity, the length of each forward movement is set to be equal and initialized to be 1 and the initial angle is set to zero.
</font> 
<br><br>

```python
# Function to draw fractals

def drawFractal(instruction, angle, x0, y0):
    x = []
    y = []

    angChange = {'+': -1*angle, '-': angle} # dictionary for angle of rotation
    
    x.append(0)
    y.append(0)
    l,x0,y0,angle = 1,0,0,0   # initialize lenth of movement, angle and starting position 
    for i in range(1,len(instruction)+1):
        if(instruction[i-1] == 'F' or instruction[i-1] == 'G'):
            dx = x[i-1] + l * np.cos(angle*np.pi/180)
            x.append(dx)
            dy = y[i-1] + l * np.sin(angle*np.pi/180)
            y.append(dy)
     
        elif(instruction[i-1] == '+'):
            angle += angChange['+'] 
            x.append(x[i-1])
            y.append(y[i-1])
            
        elif(instruction[i-1] == '-'):
            angle += angChange['-'] 
            x.append(x[i-1])
            y.append(y[i-1])


    figure, axis = plt.subplots()
    axis.plot(x,y)
    plt.axis('equal')
    plt.show()
```

<br><b> <font size="3"> Create the fractals </font> </b>


<br><font size = "3">The above function will be called everytime a fractal is to be drawn. Now let us define the fractal we want to draw. Each fractal has its own unique pattern and the pattern is built by repeating a certain sequence following distinct rule.  The functions that are used to define the fractals takes the initial pattern, and number of iterations as an argument and it constructs the pattern by using the rule at every iteration. Before going to a more complex fractals, let's look at simple geometric shapes, a square. The initial pattern for a square is F and the rule is F -> - and - -> F. After four iterations, the pattern will be F-F-F-F. Using this as an argument in the drawFractal function and 90 degrees as angle of rotation, we get a square.  </font> <br><br>

```python
drawFractal('F-F-F-F', 90, 0, 0)
```

<br><br> <font size = 3>For more complicated fractals such as Koch snowflake and sierpinski triangle, it's not easy to provide the pattern so a function that generates the patterns and is going to be used by the drawFractal function is needed. For the Kock snowflake, the starting pattern is F++F++F and the rule is, F -> F-F++F-F at each iteration.
</font> <br><br>

```python
# Function for generating Kock Snowflake patterns
# The rules are:
# start->F++F++F
# F->F-F++F-F
# angle 60

def kochSnowflake(iteration, pattern):
    if(iteration == 0):
        pattern = pattern
    else:
        for i in range(iteration):
            pattern = pattern.replace('F', 'F-F++F-F')
    
    return pattern
```

<br> <font size="3"> let's print and see what the pattern looks like after four iterations.</font><br><br>

```python
print(kochSnowflake(4,'F-F++F-F'))
```

<br> <font size="3"> The above pattern can now be drawn by calling the drawFractal function. The rotation angle is 60 degrees and initial position is at the origin whereas the the number of iterations is 6.</font><br><br>

```python
drawFractal(kochSnowflake(6, 'F++F++F'), 60, 0, 0)
```

<br><br> <font size="3"> Similarly for Sierpinski triangle, the starting pattern is F-G-G and the rule is F -> F-G+F+G-F and G -> GG at every iteration.  </font><br><br>

```python
# Function for generating Sierpinski triangle pattern. 
# The rules are:
# start->F-G-G
# F->F-G+F+G-F
# G->GG
# angle 120

def Sierpinski(iter, rules):
    if(iter == 0):
        rules = rules
    else:
        for i in range(iter):
            rules = rules.replace('G', 'GG') # this rule must be applied before the nesxt 
            rules = rules.replace('F', 'F-G+F+G-F')
    return rules
```

```python
print(Sierpinski(4,'F-G-G'))
```

<br> <font size="3"> The triangel can now be drawn by calling the drawFractal function. The rotation angle in this case 120 degrees and initial position is at the origin whereas the the number of iterations is 5.</font><br><br>

```python
drawFractal(Sierpinski(5, 'F-G-G'), 120, 0, 0)
```

<br><br> Many other complicated fractals can be generated by using similar method as long as one knows the pattern.
